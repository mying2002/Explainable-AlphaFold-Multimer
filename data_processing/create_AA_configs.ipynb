## Code to create config files for antibody-antigen complexes and monomers

```python
import pandas as pd
```

#### Raw data

```python
raw_AA_data = pd.read_excel('data/antibody_data/41598_2023_42090_MOESM3_ESM.xlsx')
```

#### Process raw data

```python
processed_AA_data = raw_AA_data
processed_AA_data = processed_AA_data.rename(columns={'ColabFold input sequence':'joined_seq', 'PDB code':"pdb_code"})
processed_AA_data['seq1'] = processed_AA_data.apply(lambda row: row.joined_seq.split(':')[0], axis=1)
processed_AA_data['seq2'] = processed_AA_data.apply(lambda row: row.joined_seq.split(':')[1], axis=1)
processed_AA_data['seq1_id'] = processed_AA_data.apply(lambda row: row.pdb_code + '_A', axis=1)
processed_AA_data['seq2_id'] = processed_AA_data.apply(lambda row: row.pdb_code + '_B', axis=1)
```

#### Grab 15 complexes

```python
n = processed_AA_data.shape[0]
selected_rows = [(i*88)//15 for i in range(15)]
selected_AA_data = processed_AA_data.iloc[selected_rows].reset_index(drop=True)

one_AA_data = selected_AA_data.iloc[[0]]
```

### Create multimer config file

```python
# multimer_AA_data = selected_AA_data[['seq1_id', 'seq2_id', 'seq1', 'seq2']]
multimer_colnames = ['seq1_id', 'seq2_id', 'seq1', 'seq2', 'which_seq', 'delete_index']
multimer_AA_data = pd.DataFrame(columns = multimer_colnames)

# loop through selected_AA_data, write rows one by one
for index, row in selected_AA_data.iterrows():
    # no deletions
    new_row_df = pd.DataFrame([[row['seq1_id'], row['seq2_id'], row['seq1'], row['seq2'], 0, -1]], columns=multimer_colnames)
    multimer_AA_data = pd.concat([multimer_AA_data, new_row_df])

    # deletions from seq1
    for i in range(len(row['seq1'])):
        new_row_df = pd.DataFrame([[row['seq1_id'], row['seq2_id'], row['seq1'], row['seq2'], 1, i]], columns=multimer_colnames)
        multimer_AA_data = pd.concat([multimer_AA_data, new_row_df])
    
    # deletions from seq2
    for j in range(len(row['seq2'])):
        new_row_df = pd.DataFrame([[row['seq1_id'], row['seq2_id'], row['seq1'], row['seq2'], 2, j]], columns=multimer_colnames)
        multimer_AA_data = pd.concat([multimer_AA_data, new_row_df])

multimer_AA_data = multimer_AA_data.reset_index(drop=True)
multimer_AA_data.insert(0, 'ArrayTaskID', (multimer_AA_data.index+1))

# create ArrayTaskID as index.
multimer_AA_data
```

### Create monomer config file
Antibodies are different from antigens

```python
monomer_colnames = ['seq_id', 'seq', 'delete_index']
monomer_AA_data = pd.DataFrame(columns = monomer_colnames)

for index, row in selected_AA_data.iterrows():
    # add in seq1
    for i in range(-1,len(row['seq1'])): # include -1 = no delete
        new_row_df = pd.DataFrame([[row['seq1_id'], row['seq1'], i]], columns=monomer_colnames)
        monomer_AA_data = pd.concat([monomer_AA_data, new_row_df])

    # add in seq2
    for j in range(-1,len(row['seq2'])): # include -1 = no delete
        new_row_df = pd.DataFrame([[row['seq2_id'], row['seq2'], j]], columns=monomer_colnames)
        monomer_AA_data = pd.concat([monomer_AA_data, new_row_df])

monomer_AA_data = monomer_AA_data.reset_index(drop=True)
monomer_AA_data.insert(0, 'ArrayTaskID', (monomer_AA_data.index+1))

monomer_AA_data
```
